# Section 2: Data Preprocessing

**Section Goals:**

In this section, our primary goals are to: *Handle Missing Data*, *remove unnecessary or repetitive features* and  *onvert Categorical String Features to Dummy Variables*. Given the large number of categorical variables in our dataset, it is essential to convert these categorical string features into dummy variables (one-hot encoding). This transformation allows us to include categorical data in statistical models and machine learning algorithms, which require numerical input. Many machine learning algorithms cannot process categorical variables directly. By converting them into dummy variables, we enable these algorithms to interpret and use the categorical data effectively. This conversion helps in capturing the impact of categorical variables on the target variable, which can be crucial for making accurate predictions.

By performing these preprocessing steps, we ensure that the dataset is clean, relevant, and ready for further analysis and modeling, which ultimately enhances the quality and effectiveness of our predictive models.

---

```python
import pandas as pd

def put_dummies(df, columns):
    df_temp = df.copy()  # Work with a copy to avoid modifying the original DataFrame
    
    for col in columns:
        if col in df_temp.columns:
            # Create dummy variables and concatenate with the original DataFrame
            dummies = pd.get_dummies(df_temp[col], prefix=col).astype(int)
            df_temp = pd.concat([df_temp, dummies], axis=1)
            
            # Drop the original categorical column
            df_temp.drop(col, axis=1, inplace=True)
    
    return df_temp
```

```python
list_dummies = ['job', 'marital', 'education', 'poutcome','contact']

for x in list_dummies:
    print(df_full[x].nunique())
```

We are checking the number of unique values in selected categorical columns to determine if they are suitable for conversion into dummy variables. Since these columns have a small number of unique values, we will convert them into dummy variables. This transformation is feasible and beneficial for preparing the data for machine learning models, as it simplifies handling categorical data.

```python
df = put_dummies(df_full,list_dummies)
# Move the 'y' column to the end of the DataFrame
df = df[[col for col in df.columns if col != 'y'] + ['y']]
```

**Introduction:**

Bias initialization adjusts the starting values of a model’s parameters to better handle class imbalances. By setting the initial bias in the final layer (e.g., \( b = \log(\frac{P}{N}) \) for binary classification), this technique helps the model better reflect the distribution of positive and negative classes from the start.

**Key Points:**

- **Purpose:**
  - To provide a balanced starting point for training in imbalanced datasets.
  
- **Benefits:**
  - Improves model performance by addressing class imbalance early in the training process.

- **Considerations:**
  - Must be applied carefully to avoid introducing additional bias.

Bias initialization can enhance model accuracy on imbalanced datasets when used appropriately.

```python
# Calculate class distribution in the training set
P = np.sum(y_train)  # Number of positive examples
N = len(y_train) - P  # Number of negative examples

# Compute the initial bias
initial_bias = np.log(P / N)

# Build the model with Dropout layers
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.5),  # Dropout layer with 50% dropout rate
    Dense(64, activation='relu'),
    Dropout(0.5),  # Dropout layer with 50% dropout rate
    Dense(1, activation='sigmoid', bias_initializer=Constant(initial_bias))
])


# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Set up early stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Train the model
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2, callbacks=[early_stopping],verbose=0)
```

### Handling Class Imbalance through Downsampling

**Introduction:**

In many real-world datasets, particularly in fields like fraud detection and spam filtering, the target classes are often imbalanced. For example, fraudulent transactions or spam emails are relatively rare compared to legitimate ones. This imbalance can lead to models that are biased towards the majority class, potentially neglecting the minority class that is critical for accurate detection. To address this issue, downsampling is employed as a strategy to balance the dataset by reducing the number of instances in the majority class, thereby enhancing the model's ability to accurately identify the minority class.

```python
# Separate the majority and minority classes
df_majority = df[df.y == 0]
df_minority = df[df.y == 1]

# Downsample the majority class
df_majority_downsampled = resample(df_majority, 
                                   replace=False,    # Sampling without replacement
                                   n_samples=len(df_minority),     # Match the number of samples to the minority class
                                   random_state=42) # Reproducibility

# Combine the minority class with the downsampled majority class
df_downsampled = pd.concat([df_minority, df_majority_downsampled])

# Check the class distribution

# Display the class distribution of the downsampled DataFrame
class_distribution = df_downsampled['y'].value_counts()

print("Class Distribution After Downsampling:")
for label, count in class_distribution.items():
    print(f"Class {label}: {count} samples")
```

```python
# Separate features and target variable
X = df_downsampled.drop(columns='y', inplace=False).values
y = df_downsampled['y'].values

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=101)

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit the scaler on the training data and transform it
X_train = scaler.fit_transform(X_train)
# Transform the test data using the same scaler
X_test = scaler.transform(X_test)
```

```python
# Build the model with Dropout layers
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.5),  # Dropout layer with 50% dropout rate
    Dense(64, activation='relu'),
    Dropout(0.5),  # Dropout layer with 50% dropout rate
    Dense(1, activation='sigmoid')
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Set up early stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Train the model
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2, callbacks=[early_stopping],verbose=0)
```

```python
losses = pd.DataFrame(history.history)

# Crear la figura
plt.figure(figsize=(10, 6))

# Graficar la pérdida de entrenamiento y validación con personalizaciones
plt.plot(losses['loss'], label='Training Loss', color='blue', linestyle='-', linewidth=2)
plt.plot(losses['val_loss'], label='Validation Loss', color='red', linestyle='--', linewidth=2)

    # Agregar título y etiquetas
plt.title('Training and Validation Loss Over Epochs', fontsize=16)
plt.xlabel('Epochs', fontsize=14)
plt.ylabel('Loss', fontsize=14)

    # Agregar cuadrícula
plt.grid(True, linestyle='--', alpha=0.7)

# Agregar leyenda
plt.legend(loc='upper right', fontsize=12)

    # Personalizar ticks
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Mostrar el gráfico
plt.show()
```

```python
# Predict on the test set
y_pred_prob = model.predict(X_test,verbose=0)
predictions = (y_pred_prob > 0.5).astype(int)

# Evaluate the performance
print(classification_report(y_test, predictions))
```

2. **Apply Downsampling Model**: If the model predicts class 1, use the model trained with downsampling. This model is specifically tuned to handle the minority class (class 1) more effectively due to the class imbalance correction provided by downsampling.

### Explanation

This model has been specifically adjusted to balance the precision and recall for class 1, making it a better choice for handling predictions of this class.

### Evaluation of the Strategy

**Advantages**:
- **Efficiency**: By using the initial model for class 0 predictions, you leverage its high accuracy and avoid unnecessary computations.
- **Focused Improvement**: By applying the downsampling model only when class 1 is predicted, you concentrate resources on improving the model's performance for the minority class.

**Disadvantages**:
- **Complexity**: Implementing a two-model system can add complexity to the workflow. It requires careful management and integration of two different models.

**Is it a Good Strategy?**

This strategy can be effective if the following conditions are met:
- **High Confidence in Initial Model**: Ensure that the initial model’s high precision for class 0 is consistently reliable across different samples.
- **Effective Downsampling Model**: Validate that the downsampling model provides substantial improvements for class 1 and performs well in practice.

Overall, this approach is sensible for scenarios where class imbalance significantly affects model performance. It leverages the strengths of both models and targets their use based on predicted class, which can lead to better overall performance and resource utilization. However, the complexity and overhead of managing two models should be carefully considered.

### Importance of This Approach in Fraud and Spam Detection

**Context:**

In fields like fraud detection and spam filtering, the nature of the data often involves significant class imbalances. In these scenarios, one class (e.g., fraudulent transactions or spam emails) is much less frequent than the other (e.g., legitimate transactions or non-spam emails). This imbalance poses unique challenges for traditional machine learning models.

**Key Points:**

1. **Class Imbalance Challenge:**
   - In fraud detection, for instance, fraudulent transactions are rare compared to legitimate ones. Similarly, spam emails are less frequent than non-spam emails. This rarity can cause standard models to become biased towards predicting the majority class, often resulting in poor performance for the minority class.

2. **Precision and Recall in Imbalanced Scenarios:**
   - **Precision** measures the accuracy of positive predictions. In fraud detection, high precision means that when the model predicts a transaction as fraudulent, it is more likely to be correct. For spam detection, high precision means that spam emails are accurately identified without misclassifying legitimate emails.
   - **Recall** measures the model's ability to identify all relevant cases. For fraud and spam detection, high recall ensures that most fraudulent transactions or spam emails are captured, minimizing the risk of missing critical cases.

3. **Tailoring Models to Class Imbalance:**
   - **Initial Model for Majority Class**: In cases where the majority class (e.g., non-fraudulent transactions or non-spam emails) has high accuracy, a well-performing initial model can be used effectively. It ensures that legitimate cases are handled with high precision, reducing false positives.
   - **Downsampling or Specialized Model for Minority Class**: For the minority class, a model that corrects for class imbalance (e.g., through downsampling) becomes crucial. By focusing on this model, you can enhance the identification of rare but critical cases, such as fraudulent transactions or spam emails, improving the overall efficacy of the detection system.

4. **Strategic Application:**
   - **High Precision for Majority Class**: Utilize models that excel at detecting the majority class to ensure that the high volume of normal cases is correctly classified, reducing the risk of false positives.
   - **Specialized Handling for Minority Class**: Apply tailored models for detecting the minority class to address the imbalances and improve recall. This ensures that the rare but important cases are identified effectively.

**Broader Implications:**

In practical applications, this approach can significantly enhance the performance of detection systems in real-world scenarios:

- **Fraud Detection Systems**: By leveraging models that handle class imbalances effectively, financial institutions can reduce the number of fraudulent transactions that go undetected. This improves security and minimizes financial losses.
- **Spam Filters**: Enhanced spam detection reduces the likelihood of legitimate emails being marked as spam, improving user experience and ensuring important communications are not missed.

**Conclusion:**

In summary, this dual-model approach is vital in fields characterized by class imbalance. It allows for high precision in the majority class while improving detection and recall for the minority class. By applying these tailored strategies, organizations can build more robust and effective systems for identifying fraud, spam, and other rare but significant events. This methodology not only enhances model performance but also ensures better overall decision-making and risk management.
